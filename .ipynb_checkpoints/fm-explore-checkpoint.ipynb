```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Import, and set to log to the console.  (See the console which is running
# Jupyter notebook for logging about HTTP requests.)
import tilemapbase
tilemapbase.start_logging()
```

```python
# Don't need if you have run before; DB file will already exist.
tilemapbase.init(create=True)
```

```python
# Use open street map
t = tilemapbase.tiles.build_OSM()
```

```python
# My current office at the University of Leeds
my_office = (-93.258133, 44.986656)

degree_range = 0.003
extent = tilemapbase.Extent.from_lonlat(my_office[0] - degree_range, my_office[0] + degree_range,
                  my_office[1] - degree_range, my_office[1] + degree_range)
extent = extent.to_aspect(1.0)
extent
```

```python
# On my desktop, DPI gets scaled by 0.75
fig, ax = plt.subplots(figsize=(8, 8), dpi=100)
ax.xaxis.set_visible(False)
ax.yaxis.set_visible(False)

plotter = tilemapbase.Plotter(extent, t, width=600)
plotter.plot(ax, t)

x, y = tilemapbase.project(*my_office)
ax.scatter(x,y, marker=".", color="black", linewidth=20)
None
```

```python
def generate_map_tiles(the_count=1):
    
    # import random
    import random
    
    # import tilemapbase
    import tilemapbase
    tilemapbase.start_logging()
    
    # Use open street map
    t = tilemapbase.tiles.build_OSM()
    
    for i in list(range(0, the_count)):
    
        # define location
        lat_top = 48.582648 
        lat_bottom = 43.600762
        lon_west = -97.002500
        lon_east = -91.823539
        
        rand_lat = random.randrange(lat_bottom * 1000000, lat_top * 1000000, 100) / 1000000
        rand_lon = random.randrange(lon_east * 1000000, lon_west * 1000000, 100) / 1000000
        
        my_location = (rand_lon, rand_lat)
        print(my_location)
        
        degree_range = 0.003
        extent = tilemapbase.Extent.from_lonlat(my_location[0] - degree_range, my_location[0] + degree_range,
                      my_location[1] - degree_range, my_location[1] + degree_range)
        extent = extent.to_aspect(1.0)
        
        # On my desktop, DPI gets scaled by 0.75
        fig, ax = plt.subplots(figsize=(8, 8), dpi=100)
        ax.xaxis.set_visible(False)
        ax.yaxis.set_visible(False)

        plotter = tilemapbase.Plotter(extent, t, width=600)
        plotter.plot(ax, t)

        x, y = tilemapbase.project(*my_location)
        ax.scatter(x,y, marker=".", color="black", linewidth=20)
        None
        
        fig.savefig("img" + str(i) + ".pdf", dpi=200)
        
        
    
```
